### Now you define your estimator with some Kernel

est.q1 modulus of the vector $\vec{q}_1$

est.q2 modulus of the vector $\vec{q}_2$

est.mu cosine of the angle between $\vec{q}_1$, $\vec{q}_2$. It is defined through: $\vec{q}_1 \cdot \vec{q}_2=|\vec{q}_1||\vec{q}_2|\mu$

### Define some constants

```python
b1 = 1.3
b20 = 0.3

shot = 10000

cg = 1.1
cs = 1. #we know this is 1.
ct = 1.1

deltac = 1.686
fnl = 1.
a1 = 1.
a2 = -17./21.

betaf = 2.*deltac*(b1-1.)
b01 = betaf*fnl

b02 = fnl**2.*4*deltac*((deltac/a1**2.)*(b20-2.*(a1+a2)*(b1-1.))-2.*(b1-1.))

b11 = fnl*(2./a1**2.)*(deltac*(b20-2*(a1+a2)*(b1-1.))-a1**2.*(b1-1.))+2.*deltac*(b1-1.)

kg = b1*cg
ks = b1*cs
kt = b1*ct
kphiphi = b1*fnl
kb20 = b20/2.
kb01 = b01
kb11 = b11
kb02 = b02

kfnl = 1.
```

```python
Ptot= b1**2.*Pnonlin+shot
Pnonlinsig = b1**2.*Pnonlin

Ptotscipy = scipy.interpolate.interp1d(K, Ptot)
Plinscipy = scipy.interpolate.interp1d(K, Plin)

Ptotsp = implemented_function('Ptot', Ptotscipy)
Plinsp = implemented_function('Plin', Plinscipy)
```

```python
est = es.Estimator(K, Ptot, Plin)

est.addF('g', 5./7.)
est.addF('s+t', 0.5*(est.q2/est.q1+est.q1/est.q2)*est.mu+(2./7.)*est.mu**2.)
#est.addF('t', (2./7.)*est.mu**2.)
est.addF('b11', 0.5*(1./M(est.q1)+1./M(est.q2)))
est.addF('b01', 0.5*est.mu*est.q1*est.q2*(1./(est.q1**2.*M(est.q2))+1./(est.q2**2.*M(est.q1))))
est.addF('phiphi', M(sp.sqrt(est.q1**2.+est.q2**2.+2*est.q1*est.q2*est.mu))*(1./M(est.q1))*(1./M(est.q2)))
#est.addF('gmod', Filter*5./7.)
#est.addF('phiphimod', Filter*M(sp.sqrt(est2.q1**2.+est2.q2**2.+2*est2.q1*est2.q2*est2.mu))*(1./M(est2.q1))*(1./M(est2.q2)))
#est.addF('smod', Filter*0.5*(est.q2/est.q1+est.q1/est.q2)*est.mu)
est.addF('b20', 1.)
est.addF('b02', 0.5*(1./(M(est.q1)*M(est.q2))))
#est.addF('fnl', b1*fnl*est.getF('phiphi')+b11*est.getF('b11')+b01*est.getF('b01'))
```

```python
lenN = len(Ngg)

dictlist = [dict() for x in range(lenN)]

cc = 1e-2

for i in range(lenN):
    dictlist[i]['gg'] = Ngg[i]
    dictlist[i]['gs'] = Ngs[i]
    dictlist[i]['sg'] = dictlist[i]['gs']
    dictlist[i]['ss'] = Nss[i]
    dictlist[i]['sphiphi'] = Nsphiphi[i]
    dictlist[i]['phiphis'] = dictlist[i]['sphiphi']
    dictlist[i]['phiphig'] = Ngphiphi[i]
    dictlist[i]['gphiphi'] = dictlist[i]['phiphig']
    dictlist[i]['phiphiphiphi'] = Nphiphiphiphi[i]
    
    
dictlist = np.array(dictlist)
```

```python
def getbiashardmatrix(Nmatrix, order, twobytwo = False):
    a, b, c = order[0], order[1], order[2]
    A = np.array([[1+Nmatrix[a+a]*0., Nmatrix[a+a]/Nmatrix[a+b], Nmatrix[a+a]/Nmatrix[a+c]], 
                  [Nmatrix[b+b]/Nmatrix[a+b], 1.+Nmatrix[b+b]*0., Nmatrix[b+b]/Nmatrix[b+c]], ##1, 3 to zero
                  [Nmatrix[c+c]/Nmatrix[a+c], Nmatrix[c+c]/Nmatrix[b+c], 1+Nmatrix[c+c]*0.]])
    if twobytwo:
        A = A[:2, :2]
    
    Ainv = np.linalg.inv(A)
    return Ainv

def getinvNelements(Nmatrix, order, twobytwo = False):
    a, b, c = order[0], order[1], order[2]
    invNels = np.array([[Nmatrix[a+a]**-1., Nmatrix[a+b]**-1., Nmatrix[a+c]**-1.],
                        [Nmatrix[a+b]**-1., Nmatrix[b+b]**-1., Nmatrix[b+c]**-1.],
                        [Nmatrix[a+c]**-1., Nmatrix[b+c]**-1., Nmatrix[c+c]**-1.]])
    if twobytwo:
        invNels = invNels[:2, :2]
    return invNels


def getbiashardmatrix2(Nmatrix, order):
    listKeys = list(itertools.product(list(order), repeat = 2))
    A = np.array([])
    for a, b in listKeys:
        A = np.append(A, Nmatrix[a+a]/Nmatrix[a+b])
    A = A.reshape((len(order), len(order)))
    Ainv = np.linalg.inv(A)
    return Ainv

def getinvNelements2(Nmatrix, order):
    listKeys = list(itertools.product(list(order), repeat = 2))
    invNels = np.array([])
    for a, b in listKeys:
        invNels = np.append(invNels, Nmatrix[a+a]/Nmatrix[a+b])
    invNels = invNels.reshape((len(order), len(order)))
    return invNels

def getcov2(order, el1, el2, Nmatrix):
    Nvec = np.array([])
    for a in order:
        Nvec = np.append(Nvec, Nmatrix[a+a])
    l = order
    i, j = l.index(el1), l.index(el2)
    Ainv = getbiashardmatrix2(Nmatrix, order)
    invN = getinvNelements2(Nmatrix, order)
    avec = Ainv[i, :]*Nvec
    bvec = Ainv[j, :]*Nvec
    temp = invN.dot(bvec)
    result = np.dot(avec.T, temp)
    return result

def getcov(order, el1, el2, Nmatrix, twobytwo = False):
    a, b, c = order[0], order[1], order[2]
    if twobytwo:
        Nvec = np.array([Nmatrix[a+a], Nmatrix[b+b]])#, Nmatrix['effeff']])
        l = order[:2] #so that I can have a runtime error
    else:
        Nvec = np.array([Nmatrix[a+a], Nmatrix[b+b], Nmatrix[c+c]])
        l = order
    i, j = l.index(el1), l.index(el2)
    Ainv = getbiashardmatrix(Nmatrix, order, twobytwo)
    invN = getinvNelements(Nmatrix, order, twobytwo)
    avec = Ainv[i, :]*Nvec
    bvec = Ainv[j, :]*Nvec
    temp = invN.dot(bvec)
    result = np.dot(avec.T, temp)
    return result
```

```python
def integrate(function, interval, opts = None):
    #mpmath.mp.pretty = True
    #result = mpmath.quad(function, interval[0], interval[1])
    result = scipy.integrate.nquad(function, interval, opts = opts)
    return result
```

```python
def _outer_integral(fa, fb, K):
        def _integrand(q, mu):
            
            modK_q = np.sqrt(K**2.+q**2.-2*K*q*mu)

            result = 2*np.pi*q**2./(2*np.pi)**3.

            result *= fa(q, K, mu)*fb(q, K, mu)
            result /= (2*Ptotscipy(q)*Ptotscipy(modK_q))

            return result
        return _integrand
```

```python
def N(K, fa, fb, minq, maxq):

        function = _outer_integral(fa, fb, K)

        options = {'limit' : 50, 'epsrel': 1e-4}
        
        opts = [options, options]
        interval = [[minq, maxq], [-1., 1.]]
        ris = integrate(function, interval, opts = opts)
        ris = ris[0]

        integral = ris

        return integral**-1
```

```python
def Integrate(K, Integrand, minq, maxq):

        function = integrand(K)

        options = {'limit' : 50, 'epsrel': 1e-4}

        opts = [options, options]
        interval = [[minq, maxq], [-1., 1.]]
        ris = integrate(function, interval, opts = opts)
        ris = ris[0]

        integral = ris

        return integral**-1
```

```python
Nvec = np.vectorize(N)
```

```python
def Fg(q1, q2, mu):
    return 5./7.

def Fphiphi(q1, q2, mu):
    q1 = float(q1)
    q2 = float(q2)
    mu = float(mu)
    return Mscipy(np.sqrt(q1**2.+q2**2.+2*q1*q2*mu))*(1./Mscipy(q1))*(1./Mscipy(q2))
```

```python
def getf(F):
    def f(q, K, mu):
        modK_q = np.sqrt(K**2.+q**2.-2*K*q*mu)
        mu_p = K**2.-q*K*mu #NOTE
        mu_p /= (K*modK_q)
        q = float(q)
        modK_q = float(modK_q)
        return 2*(F(q, K, -mu)*Plinscipy(q)+F(modK_q, K, -mu_p)*Plinscipy(modK_q))
    return f
```

```python
def getunnormalizedweight(Filter, f, K):
        def g(q, mu):
            modK_q = np.sqrt(K**2.+q**2.-2*K*q*mu)
            q = float(q)
            modK_q = float(modK_q)
            return (Filter(q, K, mu)*f(q, K, mu))/(2*Ptotscipy(q)*Ptotscipy(modK_q))
        return g
```

```python
def getintegrandfornorm(Filter, f, K):
    def norm(q, mu):
        unnormg = getunnormalizedweight(Filter, f, K)
        result = 2*np.pi*q**2./(2*np.pi)**3.
        result *= unnormg(q, mu)*f(q, K, mu)
        return result
    return norm
```

```python
def getNorm(K, Filter, f, minq, maxq):

        function = getintegrandfornorm(Filter, f, K)

        options = {'limit' : 100, 'epsrel': 1e-4}

        opts = [options, options]
        interval = [[minq, maxq], [-1., 1.]]
        
        ris = integrate(function, interval, opts = opts)
        

        integral = ris[0]

        return integral**-1
    
getNormVec = np.vectorize(getNorm)
```

```python
def getunnormvariance(Filter, f, K):
        def var(q, mu):
            result = 2*np.pi*q**2./(2*np.pi)**3.
            result *= getunnormalizedweight(Filter, f, K)(q, mu)**2.
            modK_q = np.sqrt(K**2.+q**2.-2*K*q*mu)
            q = float(q)
            modK_q = float(modK_q)
            result = (2*np.pi*q**2./(2*np.pi)**3.)*(Filter(q, K, mu)*f(q, K, mu))**2./(2*Ptotscipy(q)*Ptotscipy(modK_q))
            return result #2*(result)*(Ptotscipy(q)*Ptotscipy(modK_q))
        return var
```

```python
def getcontamination_withoutnorm(K, Filter, f, fb, minq, maxq):

        
        def function(q, mu):
            return getunnormalizedweight(Filter, f, K)(q, mu)*fb(q, K, mu)*2*np.pi*q**2./(2*np.pi)**3.

        options = {'limit' : 100, 'epsrel': 1e-4}

        opts = [options, options]
        interval = [[minq, maxq], [-1., 1.]]
        ris = integrate(function, interval, opts = opts)
        

        integral = ris

        return integral
    
getcontamination_withoutnorm_vec = np.vectorize(getcontamination_withoutnorm)
```

```python
def getvariance_withoutnorm(K, Filter, f, minq, maxq):

        function = getunnormvariance(Filter, f, K)

        options = {'limit' : 100, 'epsrel': 1e-4}

        opts = [options, options]
        interval = [[minq, maxq], [-1., 1.]]
        ris = integrate(function, interval, opts = opts)
        

        integral = ris

        return integral
    
getvariance_withoutnorm_vec = np.vectorize(getvariance_withoutnorm)
```
